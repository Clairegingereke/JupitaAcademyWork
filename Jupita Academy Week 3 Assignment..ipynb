1. Line plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

x_axis = [1,2,3,4,5]
y_axis = [10,12,8,15,7]

sns.lineplot(x = x_axis, y = y_axis, marker ='o')

plt.title("A Line Plot Chart")
plt.xlabel("x_axis")
plt.ylabel("y_axis")

plt.show()
```

2. Barplot

```python
categories = ['A', 'B', 'C', 'D']
values =  [25, 40, 15, 30]

sns.barplot(x = categories, y = values)  
              
plt.title("Bar plot")
plt.xlabel("categories")
plt.ylabel("values")
              
plt.show()              
```

3. Histogram

```python
Data =[32, 45, 50, 28, 36, 42,38,25,29,52,48,60]

sns.histplot(x= Data, bins =6)

plt.title("Histogram")
plt.xlabel("Groups")
plt.ylabel("count")

plt.show()
```

4.Scatter plot

```python
x = [3,5,7,10,12,15]
y = [25,40,30,50,35,60]

sns.scatterplot(x = x, y = y)

plt.title("Scatter plot")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

5. Pie Chart

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes =  [30, 25, 20, 15]

plt.pie (sizes, labels = labels);
```

6. Box Plot

```python
data = [45, 60, 75,300, 80,350, 90, 95, 100, 110, 120,250,200]

sns.boxplot(data);
```
